## Handling BUFR data

We load a BUFR file with radiosonde observations.

```python
import earthkit.data

fs = earthkit.data.from_source("file", "temp_10.bufr")
```

```python
len(fs)
```

We can call head(), tail() and ls() to see metadata from the header section of the BUFR messages: 

```python
fs.head()
```

to_pandas() extracts data into a Pandas dataframe (uses pdbufr under the hood). The following example shows how the get the temperature profile for a given station:

```python
df = fs.to_pandas(columns=("WMO_station_id", "data_datetime", "pressure", "airTemperature"),
    filters={"WMO_station_id": 1415})
df
```
