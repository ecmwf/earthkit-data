## GRIB: inspecting contents

```python
import earthkit.data as ekd
ekd.download_example_file("test6.grib")
```

```python
ds = ekd.from_source("file", "test6.grib")
```

### Using head(), tail() and ls()

#### head()

```python
ds.head()
```

The number of fields can be passed as an argument:

```python
ds.head(2)
```

```python
ds.head(n=2)
```

##### Keys and extra keys

We can use our own list of keys instead of the default ones.

```python
ds.head(keys="base_datetime")
```

```python
ds.head(keys=["base_datetime", "level"])
```

The default list of keys can be extended:

```python
ds.head(extra_keys="units")
```

##### Namespaces

The keys can be taken from an earthkit namespace:

```python
ds.head(namespace="time")
```

```python
ds.head(namespace=["parameter", "time"])
```

It is possible to use any [ecCodes GRIB namespace](https://confluence.ecmwf.int/display/UDOC/What+are+namespaces+-+ecCodes+GRIB+FAQ):

```python
ds.head(namespace="grib.ls")
```

#### tail()

```python
ds.tail(5)
```

#### ls()

```python
ds.ls()
```

The number of fields to list can also be specified:

```python
ds.ls(2)
```

```python
ds.ls(-2)
```

### Using describe

```python
ds.describe()
```

```python
ds.describe("t")
```

```python
ds.describe(131)
```

### Accessing metadata

```python
ds.get("vertical_type")
```

```python
ds[0].get(["level", "vertical_type"])
```

For single fields the bracket operator can also be used:

```python
ds[0].get(["grib.typeOfLevel", "grib.centre", "grib.centre"], astype=(None,str, int))
```

Namespace keys can be fetched as a dict for single fields:

```python
ds[0].namespace("parameter")
```

### Inspecting all the namespace keys for a message

```python
ds[0].dump()
```
