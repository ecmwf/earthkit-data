## GRIB: selection using metadata

We read an example GRIB file containing 18 messages and load into a fieldlist.

```python
import earthkit.data as ekd

ds = ekd.from_source("sample", "tuv_pl.grib").to_fieldlist()
len(ds)
```

### Using sel

```python
ds1 = ds.sel({"vertical.level": 500})
ds1.ls()
```

Lists are accepted:

```python
ds1 = ds.sel({"vertical.level": [500, 850]})
ds1.ls()
```

Slices can define closed intervals, so they are treated as inclusive of both the start and stop values, unlike normal Python indexing:

```python
ds1 = ds.sel({"vertical.level": slice(500, 850), "parameter.variable": "t"})
ds1.ls()
```

### Using order_by

```python
ds1 = ds.sel({"vertical.level": slice(500, 850), 
             "parameter.variable": ["u", "v"]})
```

```python
ds1.order_by("parameter.variable").ls()
```

The sorting keys can be specified as a list:

```python
ds1.order_by(["parameter.variable", "vertical.level"]).ls()
```

We can prescribe the actual order within a key. It only works when all the possible values are specified:

```python
ds1.order_by({"vertical.level": [500, 850, 700]}).ls()
```

### Combining sel and order_by

```python
ds.sel({"vertical.level": [500, 850]}).order_by(["paramater.variable"]).ls()
```
