## Retrieving data from the ADS

```python
import earthkit.data as ekd
```

### Getting GRIB data

```python
ds = ekd.from_source(
    "ads",
    "cams-global-reanalysis-eac4",
    request=dict(
        variable=["particulate_matter_10um", "particulate_matter_1um"],
        area=[50, -50, 20, 50],  # N,W,S,E
        date="2012-05-10",
        time="12:00",
    ),
)
```

```python
ds.to_fieldlist().ls()
```

### Getting NetCDF data

```python
ds1 = ekd.from_source(
    "ads",
    "cams-global-reanalysis-eac4",
    request=dict(
        variable=["particulate_matter_10um", "particulate_matter_1um"],
        area=[50, -50, 20, 50],  # N,W,S,E
        date="2012-05-10",
        time="12:00",
        format="netcdf",
    ),
)
```

```python
ds1.to_fieldlist().ls()
```
