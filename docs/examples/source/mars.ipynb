## Retrieving data from the ECMWF MARS archive

```python
import earthkit.data as ekd

ds = ekd.from_source(
    "mars",
    request=dict(
        param=["2t", "msl"],
        levtype="sfc",
        area=[50, -10, 40, 10],
        grid=[2, 2],
        date="2023-05-10",  # N,W,S,E
    ),
).to_fieldlist()
```

```python
ds.ls()
```
