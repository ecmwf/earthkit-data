## Projections

```python
import earthkit.data
from earthkit.data import remote_example_file
```

```python
efas = earthkit.data.from_source("url", remote_example_file("efas.grib"))
efas.ls()
```

```python
projection = efas[0].projection()
print(projection)
```

```python
projection.to_proj_string()
```

```python
projection.to_cartopy_crs()
```

```python
print(projection.parameters)
print(projection.globe)
```
