## GRIB: listing contents

```python
import earthkit.data as ekd
ds = ekd.from_source("sample", "test6.grib").to_fieldlist()
```

### Using ls(), head() and tail()

#### ls()

```python
ds.ls()
```

The number of fields can be passed as an argument:

```python
ds.ls(2)
```

```python
ds.ls(n=2)
```

##### Keys and extra keys

We can use our own list of keys instead of the default ones.

```python
ds.ls(keys="time.base_datetime")
```

```python
ds.ls(keys=["time.base_datetime", "vertical.level"])
```

The default list of keys can be extended with ``extra_keys``:

```python
ds.head(extra_keys="parameter.units")
```

##### Components

We can list all the keys from one or more Field component:

```python
ds.ls(component="time")
```

```python
ds.ls(component=["parameter", "time"])
```

It is possible to use any [ecCodes GRIB namespace](https://confluence.ecmwf.int/display/UDOC/What+are+namespaces+-+ecCodes+GRIB+FAQ):

```python
ds.ls(component="metadata", filter="ls")
```

#### tail()

```python
ds.tail(5)
```

#### ls()

```python
ds.ls()
```

The number of fields to list can also be specified:

```python
ds.ls(2)
```

```python
ds.ls(-2)
```

### Getting namespaces

The field namespaces can be collected as a dict via the namespace() method:

```python
ds[0].dump("parameter")
```

```python
ds[0].dump(["parameter", "time"])
```

```python
ds[0].dump()
```

With this method we can also inspect any [ecCodes GRIB namespace](https://confluence.ecmwf.int/display/UDOC/What+are+namespaces+-+ecCodes+GRIB+FAQ):

```python
ds[0].dump(part="metadata", filter="ls")
```

```python
ds[0].dump("metadata")
```

### Using dump

```python
ds[0].dump()
```

We can dump a given namespace or namespaces.

```python
ds[0].dump("parameter")
```

```python
ds[0].dump(["parameter", "time"])
```

```python
# get all the ecCodes GRIB namespaces
ds[0].dump("grib")
```

```python
# get a given ecCodes GRIB namespace
ds[0].dump("grib.ls")
```
