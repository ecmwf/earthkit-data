## Xarray engine: ensemble data

First, we get some ensemble forecast data and read it into a GRIB fieldlist.

```python
import earthkit.data as ekd
ds_fl = ekd.from_source("sample", "ens_cf_pf.grib")
```

The data contains 3 ensemble members: one control member and 2 perturbed members.

```python
ds_fl.ls()
```

```python
ds_fl.to_xarray()
```

This default can be overridden by specifying custom ``dim_roles``. E.g. to get the ensemble member number from the "perturbatioNumber" key we can use:

```python
ds_fl.to_xarray(dim_roles={"number": "perturbationNumber"})
```
