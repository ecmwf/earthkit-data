```python
# get new data
ff  = ekd.from_source("sample", "O32_t2.grib2")[0]
lat = ff.latitudes
lon = ff.longitudes
vals = ff.values
print("new geo+vals shape:", vals.shape)

# set new data
f1 = f.set(values=vals, latitudes=lat, longitudes=lon)
f1.shape, f1.latitudes.shape, f1.longitudes.shape
```

```python
# sanity check
import numpy as np
assert np.allclose(f1.values, vals)
assert np.allclose(f1.latitudes, lat)
assert np.allclose(f1.longitudes, lon)

# original metadata is kept
f1.base_datetime, f1.valid_datetime, f1.step, f1.level, f1.vertical_type
```
