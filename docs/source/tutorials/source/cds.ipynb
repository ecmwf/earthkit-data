# Retrieving data from the CDS

```python
import earthkit.data as ekd
```

### Getting GRIB data

```python
d = ekd.from_source(
    "cds",
    "reanalysis-era5-single-levels",
    request=dict(
        variable=["2t", "msl"],
        product_type="reanalysis",
        area=[50, -10, 40, 10],  # N,W,S,E
        grid=[2, 2],
        date="2012-05-10",
        time="12:00",
    ),
)
d
```

```python
d.to_fieldlist().ls()
```

### Getting NetCDF data

```python
d1 = ekd.from_source(
    "cds",
    "reanalysis-era5-single-levels",
    request=dict(
        variable=["2t", "msl"],
        product_type="reanalysis",
        area=[50, -10, 40, 10],  # N,W,S,E
        grid=[2, 2],
        date="2012-05-10",
        time="12:00",
        format="netcdf",
    ),
)
d1
```

```python
d1.to_fieldlist().ls()
```
