# Reading tar or zip archive

```python
import earthkit.data as ekd

ekd.download_example_file("test_gribs.tar")
```

```python
d = ekd.from_source("file", "test_gribs.tar")
d
```

We can convert the data into a fieldlis.

```python
fl = d.to_fieldlist()
```

```python
fl.ls()
```

```python
a = fl.sel({"vertical.level": 500})
a.ls()
```
